```python
from IPython.display import clear_output
from random import choice
from random import randrange
```

```python
def show_command_for_game():
    #the command lines as guide to user to play the game
    print("The commands for this game using the keypad pattern:")
    print("\u03327|\u03328|\u03329  <-- numbers 7,8,9 for top row")
    print("\u03324|\u03325|\u03326  <-- number 4, 5, 6 for mid row")
    print("1|2|3  <-- number 1, 2, 3 for bottom row")
    print('\n"q or quit <----- to quit game')
    print('\nMarks :\n"X" - for Comp\n"O" - for Player')
```

```python
def display_game_board():
    #the starting board to show empty marks
    show_board = []
    row1 = ['_', '|', '_', '|', '_']
    row2 = ['_', '|', '_', '|', '_']
    row3 = [' ', '|', ' ', '|', ' ']
    print("\nThe boardgame.")
    show_board.append(row1)
    show_board.append(row2)
    show_board.append(row3)
    for i in range(3):
        for _ in range(len(show_board[i])):
            print(show_board[i][_], end="")
        print("")
    return show_board
```

```python

def toss_coin(player_coin):
    #flipping coin, if the user guessed it right, the player will play first
    
    coin_sides = ['head', "tail"]
    coin = choice(coin_sides)
    
    if player_coin == coin:
        return "Player_to_start"
    else:
        return "Comp_to_start"
```

```python

def get_choice_for_coin():
    #getting the input value from player's guess
    flip_coin = ["head", "tail"]
    while True:
        
        coin_choice = input("head or tail? : ")
        if coin_choice in flip_coin:
            return coin_choice.lower()
            break
        else:
            print("Not in choices.")     
```

```python
def get_move_from_user():
    #prompting move from player based on keypad pattern
    while True:
        move_input = input("Input your move :")
        if move_input.isnumeric():
            return move_input
            break

        else:
            print("Invalid move.")
```

```python
def get_move_from_comp(movelist):
    #getting computer move based on available move on the board
    comp_move = choice(movelist)
    return comp_move
```

```python
def update_board(player_turn, game_move):
    #updating board based on player and computer input move
    u_board = display_game_board()
    game_move = int(game_move)
    game_move_list = [1,2,3,4,5,6,7,8,9]
    
    
    while True:
        game_move_list.remove(game_move)
        
        clear_output()
        #assigning underlined marks for each player
        if player_turn == "player":
            if game_move >= 4:
                mark = "\u0332" + "O"
            else:
                mark = "O"
        else:
            if game_move >= 4:
                mark = "\u0332" + "X"
            else:
                mark = "X"

        #checking move vased on value to determine what row the board should be updated
        if game_move >= 7:
            if game_move == 7:
                u_board[0][0] = mark
            elif game_move == 8:
                u_board[0][2] = mark
            else:
                u_board[0][4] = mark

        elif game_move >= 4:
            if game_move == 4:
                u_board[1][0] = mark
            elif game_move == 5:
                u_board[1][2] = mark
            else:
                u_board[1][4] = mark

        else: 
            if game_move == 1:
                u_board[2][0] = mark
            elif game_move == 2:
                u_board[2][2] = mark
            else:
                u_board[2][4] = mark

        #showing command game before showing the updated board
        show_command_for_game()
        #showing updated board after command of the game
        for i in range(3):
            for _ in range(len(u_board[i])):
                print(u_board[i][_], end="")
            print("")
            
    
        
        #checking the status of the game if it has already a winner before all possible moves are exhausted
        checked_winner_status, winner = check_for_winner(u_board)
        
        
        #checking game status
        if checked_winner_status:
            return winner
            break
        
        else:
            if len(game_move_list) == 0:
                return "Draw."
                break
            else:
                if player_turn == "player":

                    game_move = get_move_from_comp(game_move_list)
                    player_turn = "comp"
                    
                else:
                    #nested while statment to check if the user input is a possible move
                    while True:

                        game_move1 = get_move_from_user()
                        game_move1 = int(game_move1)
                        if game_move1 in game_move_list:
                            game_move = game_move1
                            player_turn =  "player"
                            
                            break
                        else:
                            print("Invalid move.")
                            continue
    
```

```python
def print_result(winning_player):
    #returning winning player
    if winning_player == "O":
        print("You win.")
    elif winning_player == "X":
        print("Computer wins.")
    else:
        print("Draw.")
```

```python
def check_for_winner(board):
    """
    Need to check for winners in these patter:
    1. 7,8,9 blocks -
    2. 7,5,3 blocks -
    3. 7,4,1 blocks -
    4. 4,5,6 blocks - 
    5. 1,5,9 blocks - 
    6. 1,2,3 blocks - 
    7. 9,6,3 blocks -
    8. 8,4,3 blocks
    """
    #checking and returning game states
    
    x = "\u0332" + "X"
    o = "\u0332" + "O"
    
    if board[0].count(x) == 3:
        return True, "X"
    elif board[0].count(o) == 3:
        return True, "O"
    elif board[0][0] == x and board[1][2] == x and board[2][4] == "X":
        return True, "X"
    elif board[0][0] == o and board[1][2] == o + board[2][4] == "O":
        return True ,"O"
    elif board[0][0] == x and board[1][0] == x and board[2][0] == "X":
        return True, "X"
    elif board[0][0] == o and board[1][0] == o and board[2][0] == "O":
        return True, "O"
    elif board[1].count(x) == 3:
        return True, "X"
    elif board[1].count(o) == 3:
        return True, "O"
    elif board[2][0] == "X" and board[1][2] == x and board[0][4] == x:
        return True, "X"
    elif board[2][0] == "O" and board[1][2] == o and board[0][4] == o:
        return True, "O"
    elif board[2].count("X") == 3:
        return True, "X"
    elif board[2].count("O") == 3:
        return True, "O"
    elif board[0][4] == x and board[1][4] == x and board[2][4] == "X":
        return True, "X"
    elif board[0][4] == o  and board[1][4] == o and board[2][4] == "O":
        return True, "O"
    elif board[0][2] == o  and board[1][2] == o and board[2][2] == "O":
        return True, "O"
    elif board[0][2] == x  and board[1][2] == x and board[2][2] == "x":
        return True, "X"
    else:
        return False, "None"
    
            
```

```python
def get_move_from_player(in_play):
    #calling the function of user move
    mlist = [1,2,3,4,5,6,7,8,9]
    if in_play.lower() == "player_to_start":
        in_play = "player"
        get_move = get_move_from_user()
        
        return in_play, get_move
    else:
        in_play = "comp"
        get_move = get_move_from_comp(mlist)
        
            
    return in_play, get_move
```

```python
def play_another_game():
    input_str = input("Another game? (Y/N) : ")
    if input_str.lower() == 'y':
        return True
    return False
```

```python
def play_game():
    clear_output()
    #show command on how to play the game
    show_command_for_game()
    #display the starting board
    display_game_board()
    #who will play first, computer or player
    player = toss_coin(get_choice_for_coin())
    #return the value of the player's move
    p_turn, move = get_move_from_player(player)
    #update the board and return the winner between computer and player
    win = update_board(p_turn, move)
    #print the name of the winner, if none, print draw
    print_result(win)
```

```python
while True:
    play_game()
    if not play_another_game():
        break
```
